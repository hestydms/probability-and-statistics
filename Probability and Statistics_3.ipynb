# Central Limit Theorem (CLT)

When we have a bunch of independent and identically distributed (i.i.d) random variables, like $X_1$, $X_2$, ..., with an average $\mu$ and a standard deviation $\sigma$, the magic of the Central Limit Theorem kicks in when we gather enough of them, i.e., when $n$ is big enough. This means that the average of these variables, $\bar{X_n}$, starts looking like a Normal distribution $N(\mu, \frac{\mu}{\sqrt{n}})$ when $n$ is big enough.

Now, let's play with sums and averages. The total sum, $S_n$, is basically the same as the sample mean times the sample size. If we multiply $\bar{X_n}$ by the number $n$, it tweaks the distribution's details, but not its main shape. So, thanks to the CLT, the sum $S_n$ also adopts an approximate $N(\mu, \frac{\mu}{\sqrt{n}})$ vibe, when $n$ is big enough.

The cool part about the CLT? It says that once we hit a certain sample size, the distribution of sample means (or sums) starts resembling the Normal shape, no matter how the population looks.

> But how big is "big enough"? Well, that's the catch – it varies.



## Example 2

Imagine you're at a café, we've got a crowd of spenders with varying habits: 40% spend 5 pounds, 40% spend 6 pounds, 19% spend 7 pounds, and 1% decide to treat themselves and their buddies by spending 30 pounds (say it's a group thing).

1. Let's dive into some simulation action. We're eyeing $\bar{X}_{30}$, the average spending of 30 customers. Can we make this look like a Normal distribution using simulation?
2. Okay, let's crank up the sample size to $\bar{X}_{100}$ now. More simulation magic. Does it give us an almost Normal distribution?
3. $\bar{X}_{300}$, time to put simulation to work again. Will it behave like a Normal distribution?

Just like you pondered about the heights of students in groups, we're doing the same with café spenders. It's all about checking if these averages, or "group means," start acting like the Normal gang.

Now let's dive into the data pool. For this population, $\mu$ = 6.03 and $\sigma$ = 2.52.

Let's take a look at the simulation results below. When we have a sample size of $n$ = 30, the distribution doesn't resemble anything close to a Normal distribution.

As we increase the sample size to $n$ = 100, the distribution gets a bit closer to Normal compared to $n$ = 30, but it's still not quite there. The spenders who splurge with 30 pounds are still causing a stir in the averages, even with a somewhat larger sample size.

However, as the sample size grows larger, those outliers start to have less influence on the mean, and the overall characteristics of the data start shining through. With samples of size 300, the distribution of sample means takes on a much more Normal-like appearance, although some hints of asymmetry remain.

1. When the sample size is $n$ = 30, the distribution of sample means does not exhibit a Normal pattern. (The noticeable "peaks" essentially correspond to samples without any individuals spending 30 pounds, samples with precisely 1 person who spent 30 pounds, samples with exactly 2 people who spent 30 pounds, and so on.)

```python
from symbulate import *
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
population = BoxModel([5, 6, 7, 30], probs = [0.4, 0.4, 0.19, 0.01])
n = 30
Xbar = RV(population ** n, mean)
xbar = Xbar.sim(10000)

xbar
```

```python
xbar.plot()
```

2. When the sample size is 100, the distribution of sample means shows less pronounced "peaks," yet it remains noticeably different from a Normal distribution.

```python
population = BoxModel([5, 6, 7, 30], probs = [0.4, 0.4, 0.19, 0.01])
n = 100
Xbar = RV(population ** n, mean)

xbar = Xbar.sim(10000)
```

```python
xbar.plot()
```

3. In the case of samples with a size of 300, the peaks have vanished, yet the distribution of sample means still exhibits a slight right skewness when compared to a Normal distribution.

```python
population = BoxModel([5, 6, 7, 30], probs = [0.4, 0.4, 0.19, 0.01])
n = 300
Xbar = RV(population ** n, mean)

xbar = Xbar.sim(10000)
```

```python
xbar.plot('hist')
Normal(6.03, 2.52 / sqrt(n)).plot()
```

The earlier example serves as a cautionary reminder. There isn't a fixed number like that determines when the Central Limit Theorem (CLT) becomes effective.

The CLT asserts that if the sample size is sufficiently large, the distribution of sample means will be approximately Normal, regardless of the underlying population distribution. However, the nature of the population distribution, which describes the distribution of individual values, plays a role in defining what constitutes a "sufficiently large" sample size.

- If the population distribution is Normal, the distribution of sample means is Normal for any sample size.
- If the population distribution is close to Normal (symmetric, single peak, light tails), smaller sample sizes can lead to an approximately Normal distribution of sample means.
- If the population distribution is far from Normal (skewed, heavy tails, outliers, multiple peaks), larger sample sizes are needed for the distribution of sample means to resemble a Normal distribution.

It's important to acknowledge that certain population distributions demand a large sample size for the CLT to apply. Yet, in numerous scenarios, the distribution of sample means approximates Normality even with small or moderate sample sizes.
